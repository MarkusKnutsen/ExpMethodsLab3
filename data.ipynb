```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Reading the data in from the files.

```python
noCol_data = sp.io.loadmat('./schlieren_data/fsss_afterNoColumns.mat')
col_data = sp.io.loadmat('./schlieren_data/fsss_afterColumns.mat')
firstWaves_data = sp.io.loadmat('./schlieren_data/fsss_firstWaves.mat')
```

Select the data to be plotted.

```python
data = firstWaves_data
title_header = ''
if data == firstWaves_data:
  title_header = 'Data from the first waves created by the wavemaker. Before any reflections.\n'
if data == col_data:
  title_header = 'Columns mounted on the beach.\n'
if data == noCol_data:
  title_header = 'No columns present. Only reflected waves and tranverse sloshing modes are present.\n'
```

Using the X and Y coordinates of the data to generate the meshgrid for the plots

```python
x = data['x']
y = data['y']
X,Y = np.meshgrid(y,x)
```

The gradient data in the meshgrid is selected from the data for a specific timestep. 

```python
gradX = data['gradX']
gradY = data['gradY']

# timestep
t = 50

# gradient for specific timestep
Z_x = gradX[:,:,t]
Z_y = gradY[:,:,t]
```

For the plots, the max and min value of the gradient is selected.

```python
max_val = max(max(Z_x.flatten()), max(Z_y.flatten()))
min_val = min(min(Z_x.flatten()), min(Z_y.flatten()))
```

Plotting the gradient in the X- and Y-direction.

```python
plt.contourf(X, Y, Z_x, 40, cmap='bwr')
plt.clim(min_val, max_val)
plt.colorbar()
plt.ylabel('x - axis')
plt.xlabel('y - axis')
plt.title(title_header + 'Free surface gradient component in the x-direction.')
plt.show()

plt.contourf(X, Y, Z_y, 40, cmap='bwr')
plt.clim(min_val, max_val)
plt.colorbar()
plt.ylabel('x - axis')
plt.xlabel('y - axis')
plt.title(title_header + 'Free surface gradient component in the y-direction.')
plt.show()
```
